```python
dataset3 = dataset3[dataset3['depth'].notna()]
print(dataset3)
```

```python
dataset3["cut"] = dataset3["cut"].astype('category')
dataset3["color"] = dataset3["color"].astype('category')
dataset3["clarity"] = dataset3["clarity"].astype('category')
dataset3.dtypes
```

```python
dataset3 = dataset3.drop(["Unnamed: 0"],axis=1)
dataset3.head()
```

```python
print("Number of rows with x == 0: {} ".format((dataset3.x==0).sum()))
print("Number of rows with y == 0: {} ".format((dataset3.y==0).sum()))
print("Number of rows with z == 0: {} ".format((dataset3.z==0).sum()))
print("Number of rows with depth == 0: {} ".format((dataset3.depth==0).sum()))
```

```python
dataset3[['x','y','z']] = dataset3[['x','y','z']].replace(0,np.NaN)
dataset3.isnull().sum()
```

```python
dataset3.dropna(inplace=True)
dataset3.isnull().sum()
```

```python
dataset3['volume'] = dataset3['x']*dataset3['y']*dataset3['z']
dataset3.drop(['x', 'y', 'z'], axis=1, inplace=True)
dataset3.head()
```

```python
sns.pairplot(dataset3)
```

```python
dataset3.plot(kind='box',figsize=(15,10),subplots=True,layout=(3,3))
plt.show()
```

```python
sns.scatterplot(data=dataset3, x=dataset3['volume'], y=dataset3['carat'])
```
